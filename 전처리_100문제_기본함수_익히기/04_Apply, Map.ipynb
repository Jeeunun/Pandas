Question 57

Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 

Unknown : N

Less than $40K : a

$40K - $60K : b

$60K - $80K : c

$80K - $120K : d

$120K +’ : e

```python
dic = {'Unknown': 'N', 'Less than $40K' : 'a' , '$40K - $60K' : 'b',
'$60K - $80K' : 'c', '$80K - $120K' : 'd', '$120K +' : 'e' }

df['newIncome'] = df['Income_Category'].map(lambda x : dic[x])
df
```

Question 58

Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 

Unknown : N

Less than $40K : a

$40K - $60K : b

$60K - $80K : c

$80K - $120K : d

$120K +’ : e

```python
def trans(x):
    if x == 'Unknown':
        return 'N'
    elif x == 'Less than $40K':
        return  'a'
    elif  x == '$40K - $60K':
        return  'b'
    elif  x == '$40K - $60K':
        return  'b'
    elif  x == '$60K - $80K':
        return  'c'
    elif  x == '$80K - $120K':
        return  'd'
    else :
        return 'e'
    
df['newIncome'] = df['Income_Category'].apply(trans)
df
```

Question 59

Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라.

 0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

```python
# 정답...
df['AgeState']  = df.Customer_Age.map(lambda x: x//10 *10)

Ans = df['AgeState'].value_counts().sort_index()
```

```python
def trans(x):
    if x < 10:
        return 0
    elif 9 < x < 20:
        return 10
    elif 19 < x < 30:
        return 20
    elif 29 < x < 40:
        return 30
    elif 39 < x < 50:
        return 40
    elif 49 < x < 60:
        return 50
    elif 59 < x < 70:
        return 60
    elif 69 < x < 80:
        return 70
    elif 79 < x < 90:
        return 80
    else:
        return 90

df['AgeState'] = df.Customer_Age.apply(trans)
df['AgeState'].value_counts()
```
