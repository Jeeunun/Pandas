### Getting & Knowing Data - DataFrame구현

#### 1) 롤 랭킹데이터

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/lol.csv", sep='\t')
df    
```

```python
# 데이터타입 확인
df.info()
```

#### 2) 제주 날씨, 인구에 따른 교통량 데이터

```python
D = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/Jeju.csv", encoding='euc-kr')
df = pd.DataFrame(D)
df
```

```python
df.info()
```

```python
# 데이터의 마지막 3개 행 출력 -> 데이터프레임 구현
last_three = {}
for c in df.columns:
    last_three[c] = df[c].tail(3)
df2 = pd.DataFrame(last_three)
df2
```

```python
for i,v in enumerate(df.dtypes):
    print(df.columns[i],v)
```

```python
# 수치형 변수의 분포(사분위, 평균, 표준편차, 최대, 최소) 확인
# 1. 수치형 변수만 추출
variable = []
for i,v in enumerate(df.dtypes):
    if v != 'object':
        variable.append(df.columns[i])

variable
```

```python
# 2. 분포 확인해보기
df[variable].describe()
```

```python
# 3. 일일이 계산해서 넣어보기
index = ['count', 'mean', 'std', 'min', '25%','50%','75%','max']
columns = variable
# range_df = pd.DataFrame(index=index, columns=columns)
```

```python
# 딕셔너리 만들기
range_dict = {}
for v in variable:
    range_dict[v] = {'count': df[v].count(), 'mean': df[v].mean(), 'std': df[v].std(), 
                    'min': df[v].min(), '25%': df[v].quantile(0.25), '50%': df[v].quantile(0.5),
                    '75%': df[v].quantile(0.75),'max': df[v].max()} 
```

```python
range_dict
```

```python
range_df = pd.DataFrame(range_dict)
range_df
```
