### Filtering & Sorting Data - DataFrame 구현

#### 1) 식당데이터

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
# quantity컬럼 값이 3인 데이터를 추출하여 첫 5행 출력 구현
data = {}
for c in df.columns:
    data[c] = df[c]

df1 = pd.DataFrame(data)
df1[df1['quantity'] == 3].head()
```

```python
# quantity, item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라
columns = ['quantity', 'item_price']
index = range(4622)

df2 = pd.DataFrame(index= index, columns=columns)
df2
```

```python
df2['quantity'] = df['quantity']
df2['item_price'] = df['item_price']
```

```python
df2
```

```python
# item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라
new_price = []
for p in df['item_price']:
    p = p.split('$')
    new_price.append(p[1])
```

```python
new_price
```

```python
df['new_price'] = new_price
```

```python
df.info()
```

```python
df['new_price'] = df['new_price'].astype('float')
```

```python
df.info()
```
